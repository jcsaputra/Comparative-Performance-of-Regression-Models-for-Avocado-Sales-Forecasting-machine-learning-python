# 1. Business Understanding

1. Penjelasan Dataset
Dataset ini merupakan pembaruan dari dataset asli yang diunggah oleh Justin, mencakup data dari 4 Januari 2015 hingga 17 Mei 2020, sedangkan dataset asli hanya mencakup periode dari 2015 hingga awal 2018. Dataset terbaru ini dibuat dengan mengunduh data untuk tahun 2018 hingga 2020 dari situs web Hass Avocado Board pada Agustus 2020. Prosesnya melibatkan pra-pemrosesan data dan penggabungannya dengan dataset milik Justin. Data baru diunduh dari halaman situs web HAB yang memungkinkan pengunduhan data untuk tahun 2017 hingga 2020 (sejauh Agustus 2020). Pengguna dapat mengunduh data baru dengan menelusuri ke bagian "Totals by PLU" dan mengklik tombol "Download 20XX Weekly Retail Volume & Price Report" (dengan "XX" merupakan dua digit terakhir tahun yang diunduh). Dataset ini kemungkinan berisi informasi tentang penjualan alpukat, termasuk volume, harga, dan mungkin variabel lainnya, memungkinkan analisis tren dan pola selama beberapa tahun.


2. Penjelasan Kolom
- Tanggal (Date): Tanggal pengamatan data penjualan alpukat.
- Harga Rata-rata (AveragePrice): Harga rata-rata untuk satu alpukat.
- Tipe (Type): Jenis alpukat, apakah konvensional atau organik.
- Tahun (Year): Tahun pengamatan data.
- Wilayah (Region): Nama kota atau daerah tempat pengamatan dilakukan.
- Total Volume: Jumlah total alpukat yang terjual.
- 4046: Jumlah total alpukat dengan kode PLU 4046 yang terjual.
- 4225: Jumlah total alpukat dengan kode PLU 4225 yang terjual.
- 4770: Jumlah total alpukat dengan kode PLU 4770 yang terjual.
- Total Bags: Jumlah total kantong yang berisi beberapa alpukat.
- Small Bags: Jumlah total kantong kecil yang berisi beberapa alpukat.
- Large Bags: Jumlah total kantong besar yang berisi beberapa alpukat.
- XLarge Bags: Jumlah total kantong sangat besar yang berisi beberapa alpukat.

Dengan kata lain, data ini memberikan informasi tentang harga, jenis, tahun, wilayah, dan volume penjualan alpukat. Selain itu, kita juga dapat melihat rincian penjualan berdasarkan jenis kemasan, seperti kantong kecil, kantong besar, dan sebagainya. Jumlah alpukat dengan kode PLU tertentu (4046, 4225, dan 4770) juga dicantumkan, memberikan informasi lebih lanjut tentang variasi jenis alpukat yang terjual.

Tujuan dari analisis data adalah memahami faktor-faktor yang mempengaruhi penjualan alpukat Hass untuk membantu pengambilan keputusan dan strategi pemasaran

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

### 1. Multiple Regression

```python
X = avocado[['AveragePrice', '4046', '4225', '4770', 'type', 'year']]
y = avocado['Total Volume']
y = y.values.reshape(-1, 1)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
print('Coefficients: ',model.coef_)
print('Intercept: ',model.intercept_)
```

```python
y_pred = model.predict(X_test)
print('Predicted: ', y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae_multi = mean_absolute_error(y_test, y_pred)
mse_multi = mean_squared_error(y_test, y_pred)
r2_multi = r2_score(y_test, y_pred)

print("Mean Absolute Error: %.2f" % mae_multi)
print("Mean Squared Error: %.2f" % mse_multi)
print("R-squared:", r2_multi)
```

```python
y_test_inverse = scaler.inverse_transform(y_test.reshape(-1,1))
y_pred_inverse = scaler.inverse_transform(y_pred.reshape(-1,1))
```

```python
line_color = 'red'
ax = sns.regplot(x=y_test_inverse, y=y_pred_inverse, scatter_kws={'color': 'blue'}, line_kws={'color': line_color})
plt.show()
```

### 3. Random Forest Regression

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

X = avocado[['AveragePrice', '4046', '4225', '4770', 'type', 'year']]
y = avocado['Total Volume']
y = y.values.reshape(-1, 1)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

X_train
```

Di atas adalah matriks fitur (X_train) yang digunakan untuk melatih model RandomForestRegressor. Setiap baris dalam matriks berhubungan dengan titik data tertentu, dan setiap kolom mewakili fitur yang berbeda. Fitur-fitur tersebut telah distandarisasi atau dinormalisasi, karena nilainya bervariasi di sekitar nol. Enam kolom kemungkinan besar berhubungan dengan enam fitur berbeda yang digunakan untuk prediksi.

```python
# Membuat model Random Forest
Rand_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model dengan data latih
Rand_model.fit(X_train, y_train.ravel())

# Memprediksi data uji
y_pred = Rand_model.predict(X_test)

y_pred
```

Di atas adalah prediksi untuk variabel target berdasarkan fitur input. Setiap nilai sesuai dengan hasil prediksi untuk titik data tertentu dalam kumpulan data. Sebagai contoh, nilai pertama (-0.38995504) adalah hasil prediksi untuk titik data pertama, nilai kedua (1.95701376) sesuai dengan titik data kedua, dan seterusnya. Pola ini terus berlanjut di seluruh array, dengan setiap nilai mewakili prediksi untuk titik data tertentu.

```python
importances = Rand_model.feature_importances_
for feature, importance in zip(X.columns, importances):
    print(f"{feature}: {importance}")
```

Total volume penjualan alpukat terutama dipengaruhi oleh jumlah alpukat Hass besar yang terjual (4225), yang menyumbang 60,78% dari variasi penjualan. Di bawahnya adalah jumlah total alpukat Hass kecil/sedang yang terjual (4046), yang menyumbang 27,15% terhadap variasi. Tahun pengamatan adalah fitur terpenting ketiga (nilai penting fitur 0,0628), yang menunjukkan tren temporal atau musiman dalam penjualan, yang mungkin dipengaruhi oleh faktor-faktor seperti penawaran, permintaan, harga, cuaca, atau preferensi konsumen. Harga rata-rata alpukat (AveragePrice) adalah fitur terpenting keempat, menjelaskan 2,77% variasi, menunjukkan dampak yang relatif lebih rendah dibandingkan dengan fitur lainnya. Jumlah total alpukat Hass ekstra besar yang terjual (4770) merupakan fitur terpenting kelima, dengan kontribusi sebesar 2,12%. Jenis alpukat (konvensional atau organik) adalah fitur yang paling tidak berpengaruh, menjelaskan hanya 0,90% dari variasi, menunjukkan pengaruh yang dapat diabaikan terhadap total penjualan dibandingkan dengan faktor lainnya.

```python
plt.scatter(y_test, y_pred, color='blue', label='Actual vs. Predicted')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2, label='Perfect Prediction')
plt.xlabel('Actual Total Volume')
plt.ylabel('Predicted Total Volume')
plt.title('Scatter Plot Random Forest Regression: Actual vs. Predicted')
plt.legend()
plt.show()
```
